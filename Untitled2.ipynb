```python
import pandas as pd
from functools import reduce

# Clean Task and Qual to avoid hidden issues
df1['Task'] = df1['Task'].str.strip()
df1['Qual'] = df1['Qual'].str.strip()
df1['ID'] = df1['ID'].astype(int)

# Define the weeks to extract
weeks_to_extract = [0, 6, 12, 16]

# Container for pivoted data
pivoted_psr = []

for week in weeks_to_extract:
    temp = df1[df1['Week'] == week].copy()
    
    # Pivot: one row per ID, one column per Task+Qual
    pivot = temp.pivot_table(index='ID', columns=['Task', 'Qual'], values='PS R')
    
    # Rename columns to indicate the week
    pivot.columns = [f'PSR_W{week}_{task}_{qual}' for task, qual in pivot.columns]
    
    # Add to list
    pivoted_psr.append(pivot)

# Merge all weeks using OUTER join to retain all IDs
psr_combined = reduce(
    lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='outer'),
    pivoted_psr
)

# Optional: reset index if you want 'ID' as a column
psr_combined.reset_index(inplace=True)

# Optional: fill missing values (if you want to treat missing scores as 0)
# psr_combined.fillna(0, inplace=True)

#  Done
psr_combined.head()
print(list(psr_combined))
```

```python
import pandas as pd
from functools import reduce

# Clean Task, Qual, and ID just once
df3['Task'] = df3['Task'].str.strip()
df3['Qual'] = df3['Qual'].str.strip()
df3['ID'] = df3['ID'].astype(int)

# Define weeks and gait-related variables
weeks_to_extract = [0, 6, 12, 16]

gait_variables = [
    'Gait symmetry', 'Step length', 'Step length left', 'Step length right',
    'Step time', 'Step time left', 'Step time right',
    'Step length var', 'Step time var',
    'Step length asym', 'Step time asym',
    'Step velocity',
    'Step count lap 1', 'Step count lap 2', 'Step count lap 3', 'Step count lap 4'
]

# Master container for all pivoted variables
all_pivoted = []

for var in gait_variables:
    pivoted_list = []
    for week in weeks_to_extract:
        temp = df3[df3['Week'] == week].copy()
        pivot = temp.pivot_table(index='ID', columns=['Task', 'Qual'], values=var)
        
        # Rename columns to reflect variable and week
        pivot.columns = [f'{var.replace(" ", "").replace("-", "").replace("/", "")}_W{week}_{task}_{qual}' for task, qual in pivot.columns]
        pivoted_list.append(pivot)
    
    # Merge all weeks for this variable
    merged_var = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), pivoted_list)
    
    all_pivoted.append(merged_var)

# Merge all variables together on ID
final_combined_gait = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), all_pivoted)

# Optional: reset index
final_combined_gait.reset_index(inplace=True)

#  Preview the result
final_combined_gait.head()
#print(list(final_combined_gait))
```

```python
import pandas as pd
import numpy as np
from merf import MERF
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score

# -----------------------------
# Helpers
# -----------------------------
def make_weekly_cols(df):
    return sorted(
        [c for c in df.columns if c.startswith("WEEK") and c.endswith("_AVERAGE")],
        key=lambda x: int(x.replace("WEEK", "").replace("_AVERAGE",""))
    )

def add_bmi_and_cat(df):
    if "BMI" not in df.columns or "BMI_cat" not in df.columns:
        if not {"Height.cm","Weight.kg"}.issubset(df.columns):
            raise ValueError("Height.cm and Weight.kg required to compute BMI/BMI_cat.")
        h_m = df["Height.cm"] / 100.0
        df["BMI"] = df["Weight.kg"] / (h_m ** 2)
        df["BMI_cat"] = pd.cut(
            df["BMI"], bins=[-np.inf, 18.5, 25, np.inf],
            labels=["Underweight", "Normal", "Obese"]
        )
    return df

def build_X(df, weekly_cols, use_bmi_cat=True, extra_numeric=None):
    # WEEKS_ATTENDED
    if "WEEKS_ATTENDED" not in df.columns:
        df["WEEKS_ATTENDED"] = (df[weekly_cols] > 0).sum(axis=1)

    cols = weekly_cols + ["WEEKS_ATTENDED"]

    # Optional numeric covariates (kept if present)
    extra_numeric = extra_numeric or []
    cols += [c for c in extra_numeric if c in df.columns]

    X = df[cols].copy()

    # Add BMI_cat one-hot
    if use_bmi_cat:
        X = pd.concat([X, pd.get_dummies(df["BMI_cat"], prefix="BMI", dummy_na=True)], axis=1)

    return X

def fit_merf_return_scores(X, y, clusters, test_size=0.2, random_state=42):
    Z = np.ones((X.shape[0], 1))  # random intercept
    X_train, X_test, Z_train, Z_test, g_train, g_test, y_train, y_test = train_test_split(
        X, Z, clusters, y, test_size=test_size, random_state=random_state
    )
    imp = SimpleImputer(strategy="median")
    X_train_imp = imp.fit_transform(X_train)
    X_test_imp = imp.transform(X_test)

    rf = RandomForestRegressor(n_estimators=500, random_state=random_state, n_jobs=-1)
    merf = MERF(rf)
    merf.fit(X_train_imp, Z_train, g_train, y_train)
    y_pred = merf.predict(X_test_imp, Z_test, g_test)

    rmse = float(np.sqrt(mean_squared_error(y_test, y_pred)))
    r2 = float(r2_score(y_test, y_pred))
    return rmse, r2, merf, rf, (X_train, X_test, y_train, y_test)

# -----------------------------
# Prepare data
# -----------------------------
# Target and filter rows with target present
target = "PSR_W12_Firm_EC"
y_full = df[target]
mask = y_full.notna()
dfm = df.loc[mask].copy()
y = y_full.loc[mask].copy()

# Ensure BMI & BMI_cat exist
dfm = add_bmi_and_cat(dfm)

# Weekly averages
weekly_avg_cols = make_weekly_cols(dfm)

# Optional extra numeric covariates (keep if present)
extra_numeric = ["Age.years", "Height.cm", "Weight.kg"]

# Grouping
clusters = dfm["ID"]

# -----------------------------
# Global MERF (with BMI_cat)
# -----------------------------
X_global = build_X(dfm, weekly_avg_cols, use_bmi_cat=True)
rmse_g, r2_g, merf_g, rf_g, (Xtr, Xte, ytr, yte) = fit_merf_return_scores(X_global, y, clusters)

print("\n=== Global MERF with BMI_cat ===")
print(f"RMSE: {rmse_g:.4f}")
print(f"R:   {r2_g:.4f}")

# --- Feature importance (from the RF fixed-effects learner) ---
importances = pd.Series(rf_g.feature_importances_, index=X_global.columns).sort_values(ascending=False)
print("\nTop 20 features:")
print(importances.head(20).to_string())
```

```python
import numpy as np
import pandas as pd
from merf import MERF
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score

# --- CONFIG ---
target = "Gaitsymmetry_W12_Walk_HT"

# Use the weeks you kept
kept_weeks = [
    'WEEK1_AVERAGE','WEEK2_AVERAGE','WEEK3_AVERAGE','WEEK4_AVERAGE','WEEK5_AVERAGE',
    'WEEK7_AVERAGE','WEEK8_AVERAGE','WEEK9_AVERAGE','WEEK10_AVERAGE','WEEK11_AVERAGE'
]

# --- Select PSR variables (exclude Week 12) ---
psr_cols = [c for c in df.columns if c.startswith("PSR_") and "_W12_" and "_W16_" not in c]

# --- Build modeling frame ---
need_cols = [target, "ID"] + kept_weeks #+ psr_cols
dfm = df[need_cols].copy()

# Preserve your semantics: 0 = not attended. Fill only real NaNs in weekly with 0.
dfm[kept_weeks] = dfm[kept_weeks].fillna(0)

# Drop rows with missing target or ID
dfm = dfm.dropna(subset=[target, "ID"]).reset_index(drop=True)

# --- Fixed-effects matrix X (weekly + PSR) ---
X = dfm[kept_weeks].copy()

# --- Random-effects design Z: intercept + random slopes for each kept week ---
# Standardize week columns for stable random-slope estimation (mean 0, var ~1)
Z_weeks = dfm[kept_weeks].astype(float)
Z_weeks = (Z_weeks - Z_weeks.mean(axis=0)) / (Z_weeks.std(axis=0).replace(0, 1.0))

# Z = [1, z_week1, z_week2, ...]
Z = np.column_stack([np.ones((len(dfm), 1)), Z_weeks.values])

# Keep names for reference
z_colnames = ["(Intercept)"] + [f"RS_{c}" for c in kept_weeks]

# --- Group labels and target ---
groups = dfm["ID"]
y = dfm[target]

# --- Train/test split ---
X_train, X_test, Z_train, Z_test, g_train, g_test, y_train, y_test = train_test_split(
    X, Z, groups, y, test_size=0.2, random_state=42
)

# --- Impute any remaining NaNs in X (median). Z is already numeric/standardized. ---
imp = SimpleImputer(strategy="median")
X_train_imp = imp.fit_transform(X_train)
X_test_imp  = imp.transform(X_test)

# --- MERF with Random Forest fixed effects ---
rf = RandomForestRegressor(
    n_estimators=600,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)
merf = MERF(rf)

# Fit
merf.fit(X_train_imp, Z_train, g_train, y_train)

# Predict & evaluate
y_pred = merf.predict(X_test_imp, Z_test, g_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE: {rmse:.4f}")
print(f"Test R:   {r2:.4f}")

# --- Fixed-effects feature importance (Random Forest) ---
fe_importance = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 20 fixed-effect features:")
print(fe_importance.head(20).to_string())

# --- FYI: Random-effects structure you fit ---
print("\nRandom effects design (Z) columns:")
print(z_colnames)
```
