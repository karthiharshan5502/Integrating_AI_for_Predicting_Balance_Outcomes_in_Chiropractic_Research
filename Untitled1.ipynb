```python
import pandas as pd
from functools import reduce

# Clean Task, Qual, and ID just once
df3['Task'] = df3['Task'].str.strip()
df3['Qual'] = df3['Qual'].str.strip()
df3['ID'] = df3['ID'].astype(int)

# Define weeks and gait-related variables
weeks_to_extract = [0, 6, 12, 16]

gait_variables = [
    'Gait symmetry', 'Step length', 'Step length left', 'Step length right',
    'Step time', 'Step time left', 'Step time right',
    'Step length var', 'Step time var',
    'Step length asym', 'Step time asym',
    'Step velocity',
    'Step count lap 1', 'Step count lap 2', 'Step count lap 3', 'Step count lap 4'
]

# Master container for all pivoted variables
all_pivoted = []

for var in gait_variables:
    pivoted_list = []
    for week in weeks_to_extract:
        temp = df3[df3['Week'] == week].copy()
        pivot = temp.pivot_table(index='ID', columns=['Task', 'Qual'], values=var)
        
        # Rename columns to reflect variable and week
        pivot.columns = [f'{var.replace(" ", "").replace("-", "").replace("/", "")}_W{week}_{task}_{qual}' for task, qual in pivot.columns]
        pivoted_list.append(pivot)
    
    # Merge all weeks for this variable
    merged_var = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), pivoted_list)
    
    all_pivoted.append(merged_var)

# Merge all variables together on ID
final_combined_gait = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), all_pivoted)

# Optional: reset index
final_combined_gait.reset_index(inplace=True)

#  Preview the result
final_combined_gait.head()
#print(list(final_combined_gait))
```

```python
import pandas as pd
import numpy as np
from merf import MERF
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# -----------------------------
# Helpers
# -----------------------------
def make_weekly_cols(df):
    return sorted(
        [c for c in df.columns if c.startswith("WEEK") and c.endswith("_AVERAGE")],
        key=lambda x: int(x.replace("WEEK", "").replace("_AVERAGE",""))
    )

def add_bmi_and_cat(df):
    # Compute BMI & BMI_cat if not present (requires Height.cm, Weight.kg)
    if "BMI" not in df.columns or "BMI_cat" not in df.columns:
        if not {"Height.cm","Weight.kg"}.issubset(df.columns):
            raise ValueError("Height.cm and Weight.kg required to compute BMI/BMI_cat.")
        h_m = df["Height.cm"] / 100.0
        df["BMI"] = df["Weight.kg"] / (h_m ** 2)
        df["BMI_cat"] = pd.cut(
            df["BMI"], bins=[-np.inf, 18.5, 25, np.inf],
            labels=["Underweight", "Normal", "Obese"]
        )
    return df

def build_X(df, weekly_cols, use_bmi_cat=False, extra_numeric=None):
    # WEEKS_ATTENDED
    if "WEEKS_ATTENDED" not in df.columns:
        df["WEEKS_ATTENDED"] = (df[weekly_cols] > 0).sum(axis=1)

    # Base predictors
    cols = weekly_cols + ["WEEKS_ATTENDED"]

    # Add optional extra numeric covariates if present
    extra_numeric = extra_numeric or []
    cols += [c for c in extra_numeric if c in df.columns]

    X = df[cols].copy()

    # Optionally add BMI_cat one-hot (global model)
    if use_bmi_cat:
        dummies = pd.get_dummies(df["BMI_cat"], prefix="BMI", dummy_na=True)
        X = pd.concat([X, dummies], axis=1)

    return X

def fit_merf_return_scores(X, y, clusters, test_size=0.2, random_state=42):
    Z = np.ones((X.shape[0], 1))
    X_train, X_test, Z_train, Z_test, g_train, g_test, y_train, y_test = train_test_split(
        X, Z, clusters, y, test_size=test_size, random_state=random_state
    )
    imp = SimpleImputer(strategy="median")
    X_train_imp = imp.fit_transform(X_train)
    X_test_imp = imp.transform(X_test)

    rf = RandomForestRegressor(n_estimators=500, random_state=random_state, n_jobs=-1)
    merf = MERF(rf)
    merf.fit(X_train_imp, Z_train, g_train, y_train)
    y_pred = merf.predict(X_test_imp, Z_test, g_test)

    rmse = float(np.sqrt(mean_squared_error(y_test, y_pred)))
    r2 = float(r2_score(y_test, y_pred))
    return rmse, r2, merf, rf, (X_train, X_test, y_train, y_test)

# -----------------------------
# Prepare data
# -----------------------------
# Target and filter rows with target present
y_full = df["Gaitsymmetry_W12_Walk_HT"]
mask = y_full.notna()
dfm = df.loc[mask].copy()
y = y_full.loc[mask].copy()

# Ensure BMI & BMI_cat exist (needs Height.cm + Weight.kg)
dfm = add_bmi_and_cat(dfm)

# Weekly average columns
weekly_avg_cols = make_weekly_cols(dfm)

# Optional extra numeric covariates (include if available)
extra_numeric = ["Age.years", "Height.cm", "Weight.kg"]

# Grouping
clusters = dfm["ID"]

# -----------------------------
# A) Global MERF (with BMI_cat)
# -----------------------------
X_global = build_X(dfm, weekly_avg_cols, use_bmi_cat=True, extra_numeric=extra_numeric)
rmse_g, r2_g, merf_g, rf_g, _ = fit_merf_return_scores(X_global, y, clusters)

# -----------------------------
# B) Separate MERFs per BMI_cat
# -----------------------------
results = []
for cat in ["Underweight", "Normal", "Obese"]:
    sub = dfm[dfm["BMI_cat"] == cat].copy()
    if len(sub) < 10:
        # too few rows to split; skip but record
        results.append({
            "Model": f"MERF_{cat}",
            "N": len(sub),
            "RMSE": np.nan,
            "R2": np.nan
        })
        continue

    X_cat = build_X(sub, weekly_avg_cols, use_bmi_cat=False, extra_numeric=extra_numeric)
    rmse_c, r2_c, merf_c, rf_c, _ = fit_merf_return_scores(X_cat, sub["Gaitsymmetry_W12_Walk_HT"], sub["ID"])
    results.append({
        "Model": f"MERF_{cat}",
        "N": len(sub),
        "RMSE": rmse_c,
        "R2": r2_c
    })

# -----------------------------
# Summary table
# -----------------------------
summary_rows = [{"Model": "MERF_Global_with_BMIcat", "N": len(dfm), "RMSE": rmse_g, "R2": r2_g}]
summary_rows.extend(results)
summary_df = pd.DataFrame(summary_rows)

print("\n=== Performance Comparison ===")
print(summary_df.to_string(index=False))
```

```python
import numpy as np
import pandas as pd
from merf import MERF
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import train_test_split, GroupKFold
from sklearn.metrics import mean_squared_error, r2_score

# ---------- helpers ----------
def weekly_cols(df):
    return sorted(
        [c for c in df.columns if c.startswith("WEEK") and c.endswith("_AVERAGE")],
        key=lambda x: int(x.replace("WEEK","").replace("_AVERAGE",""))
    )

def ensure_bmi_cat(df):
    if "BMI_cat" not in df.columns:
        if not {"Height.cm","Weight.kg"}.issubset(df.columns):
            raise ValueError("Need Height.cm and Weight.kg to compute BMI_cat.")
        h_m = df["Height.cm"]/100.0
        df["BMI"] = df["Weight.kg"] / (h_m**2)
        df["BMI_cat"] = pd.cut(
            df["BMI"],
            bins=[-np.inf, 18.5, 25, np.inf],
            labels=["Underweight","Normal","Obese"]
        )
    return df

def build_X(df):
    wcols = weekly_cols(df)
    df = df.copy()
    df["WEEKS_ATTENDED"] = (df[wcols] > 0).sum(axis=1)

    X_base = df[wcols + ["WEEKS_ATTENDED"]]
    X_bmi  = pd.get_dummies(df["BMI_cat"], prefix="BMI", dummy_na=True)

    extra = [c for c in ["Age.years","Height.cm","Weight.kg"] if c in df.columns]
    X_extra = df[extra] if extra else None

    parts = [X_base, X_bmi] + ([X_extra] if X_extra is not None else [])
    X = pd.concat(parts, axis=1)
    return X, wcols

def eval_merf_params(X, Z, y, groups, params, n_splits=5, random_state=42):
    gkf = GroupKFold(n_splits=n_splits)
    rmses = []
    for tr_idx, va_idx in gkf.split(X, y, groups):
        X_tr, X_va = X.iloc[tr_idx], X.iloc[va_idx]
        Z_tr, Z_va = Z[tr_idx], Z[va_idx]
        g_tr, g_va = groups.iloc[tr_idx], groups.iloc[va_idx]
        y_tr, y_va = y.iloc[tr_idx], y.iloc[va_idx]

        gbr = HistGradientBoostingRegressor(
            learning_rate=params["learning_rate"],
            max_iter=params["max_iter"],
            max_leaf_nodes=params["max_leaf_nodes"],
            min_samples_leaf=params["min_samples_leaf"],
            l2_regularization=params["l2_regularization"],
            max_bins=params["max_bins"],
            early_stopping=True,
            validation_fraction=0.1,
            random_state=random_state
        )
        merf = MERF(gbr)
        merf.fit(X_tr, Z_tr, g_tr, y_tr)
        y_hat = merf.predict(X_va, Z_va, g_va)
        rmses.append(np.sqrt(mean_squared_error(y_va, y_hat)))
    return float(np.mean(rmses))

# ---------- data ----------
y_full = df["Gaitsymmetry_W12_Walk_HT"]
mask = y_full.notna()
dfm = df.loc[mask].copy()
y = y_full.loc[mask].copy()
dfm = ensure_bmi_cat(dfm)

X, wcols = build_X(dfm)
clusters = dfm["ID"].copy()
Z = np.ones((X.shape[0], 1))

# train/test split (by rows, but we keep clusters for MERF)
X_tr, X_te, Z_tr, Z_te, g_tr, g_te, y_tr, y_te = train_test_split(
    X, Z, clusters, y, test_size=0.2, random_state=42
)

# ---------- hyperparameter search (small grid; expand if you like) ----------
param_grid = [
    {"learning_rate": lr, "max_iter": it, "max_leaf_nodes": mln,
     "min_samples_leaf": msl, "l2_regularization": l2, "max_bins": mb}
    for lr in [0.03, 0.06, 0.1]
    for it in [400, 600, 900]
    for mln in [31, 63, 127]
    for msl in [10, 20, 50]
    for l2 in [0.0, 0.5, 1.0]
    for mb in [255]
]

best_score = np.inf
best_params = None

# Use only the training split for CV to avoid peeking at test
X_cv = X_tr.reset_index(drop=True)
Z_cv = Z_tr
y_cv = y_tr.reset_index(drop=True)
g_cv = g_tr.reset_index(drop=True)

# Light random subset of the grid to keep time reasonable
rng = np.random.RandomState(42)
subset_idx = rng.choice(len(param_grid), size=min(30, len(param_grid)), replace=False)

for i in subset_idx:
    params = param_grid[i]
    cv_rmse = eval_merf_params(X_cv, Z_cv, y_cv, g_cv, params, n_splits=5, random_state=42)
    if cv_rmse < best_score:
        best_score = cv_rmse
        best_params = params

print("Best CV RMSE:", round(best_score, 4))
print("Best params:", best_params)

# ---------- fit final MERF on train with best params ----------
gbr = HistGradientBoostingRegressor(
    learning_rate=best_params["learning_rate"],
    max_iter=best_params["max_iter"],
    max_leaf_nodes=best_params["max_leaf_nodes"],
    min_samples_leaf=best_params["min_samples_leaf"],
    l2_regularization=best_params["l2_regularization"],
    max_bins=best_params["max_bins"],
    early_stopping=True,
    validation_fraction=0.1,
    random_state=42
)
merf = MERF(gbr)
merf.fit(X_tr, Z_tr, g_tr, y_tr)

# ---------- evaluate on held-out test ----------
y_pred = merf.predict(X_te, Z_te, g_te)
rmse = np.sqrt(mean_squared_error(y_te, y_pred))
r2 = r2_score(y_te, y_pred)
print(f"Test RMSE: {rmse:.4f}")
print(f"Test R:   {r2:.4f}")

# ---------- feature importances (from tuned fixed-effects learner) ----------
# HGBR exposes feature_importances_
imp = pd.Series(gbr.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 20 features:")
print(imp.head(20).to_string())
```

```python
import numpy as np
import pandas as pd
from merf import MERF
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score

df0 = final_merged.copy()

# --- Target ---
target = "Gaitsymmetry_W12_Walk_HT"
df0 = df0.loc[df0[target].notna()].copy()
y = df0[target]

all_cols = df0.columns

# --- Weekly averages: include ALL weeks (e.g., WEEK1_AVERAGE ... WEEK20_AVERAGE) ---
weekly_avg_cols = sorted(
    [c for c in all_cols if c.startswith("WEEK") and c.endswith("_AVERAGE")],
    key=lambda x: int(x.replace("WEEK","").replace("_AVERAGE",""))
)

# --- PSR variables: include ALL except Week 12 ---
psr_cols = [
    c for c in all_cols
    if c.startswith("PSR_") and ("_W12_" not in c)
]

# --- Gait symmetry predictors: W0/W6 only, exclude W12 and exclude any other Walk_HT (to avoid leakage) ---
gait_cols = [
    c for c in all_cols
    if c.startswith("Gaitsymmetry_")
       and (("_W0_" in c) or ("_W6_" in c))
       and ("_W12_" not in c)
       and (c != target)
       and (not c.endswith("_Walk_HT"))
]

# --- Step velocity predictors: W0/W6 only (no W12) ---
velocity_cols = [
    c for c in all_cols
    if c.startswith("Stepvelocity_")
       and (("_W0_" in c) or ("_W6_" in c))
       and ("_W12_" not in c)
]

# --- Combine predictors (no other predictors like FIRST/MID/LAST/SLOPE*/AGE/GROUP) ---
predictors = weekly_avg_cols + psr_cols + gait_cols + velocity_cols

# Subset & keep ID for grouping
use_cols = predictors + [target, "ID"]
dfm = df0[use_cols].copy()

# One-hot encode NOTHING here (you asked to exclude AGE/GROUP etc.)
# If any remaining categorical sneaked in, get_dummies safely handles them:
df_encoded = pd.get_dummies(dfm, drop_first=True)

# Final X/y and groups
X = df_encoded.drop(columns=[target])
y = df_encoded[target]
clusters = df_encoded["ID"]

# Remove ID from fixed-effect features
X = X.drop(columns=["ID"])

# --- MERF setup ---
Z = np.ones((X.shape[0], 1))  # random intercept

# Train/test split
X_train, X_test, Z_train, Z_test, g_train, g_test, y_train, y_test = train_test_split(
    X, Z, clusters, y, test_size=0.2, random_state=42
)

# Impute only real NaNs (keep your meaningful zeros)
imp = SimpleImputer(strategy="median")
X_train_imp = imp.fit_transform(X_train)
X_test_imp = imp.transform(X_test)

# Mixed-Effects Random Forest
rf = RandomForestRegressor(n_estimators=500, random_state=42, n_jobs=-1)
merf = MERF(rf)
merf.fit(X_train_imp, Z_train, g_train, y_train)

# Predict & evaluate
y_pred = merf.predict(X_test_imp, Z_test, g_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.4f}")
print(f"R:   {r2:.4f}")

# Feature importance
feat_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 20 features:")
print(feat_imp.head(20).to_string())
```
